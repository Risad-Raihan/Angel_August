```python
from openpyxl.styles.colors import WHITE, RGB

__old_rgb_set__ = RGB.__set__


def __rgb_set_fixed__(self, instance, value):
    try:
        __old_rgb_set__(self, instance, value)
    except ValueError as e:
        if e.args[0] == 'Colors must be aRGB hex values':
            __old_rgb_set__(self, instance, WHITE)  # Change default color here


RGB.__set__ = __rgb_set_fixed__
```

```python
import numpy as np

# Ensure 'date' and 'Joining Date' are datetime format (if not already)
df['date'] = pd.to_datetime(df['date'])
df['Joining Date'] = pd.to_datetime(df['Joining Date'])

# Calculate difference in days (ensure integer division)
df['tenure_days'] = (df['date'] - df['Joining Date']).dt.days  # dt.days for difference in days

# Convert tenure days to months (assuming 30 days per month)
df['tenure'] = df['tenure_days'] / 30
```

# Input for Angel condition

```python
tenure = 24
regularity = 2
amount = 1200
```

# Applying Angel Condition

```python
angel = df.loc[(df['Regularity'] >= regularity) & (df['tenure'] >= tenure ) & ((((df['Month_1_Customer_Sales'] >= amount) & (df['Month_2_Customer_Sales'] >= amount)) |
                ((df['Month_2_Customer_Sales'] >= amount) & (df['Month_3_Customer_Sales'] >= amount)) |
                ((df['Month_1_Customer_Sales'] >= amount) & (df['Month_3_Customer_Sales'] >= amount))))]
```
