```python
# Select the columns to keep
columns_to_keep = [
    'Customer Code',
    'Maritial Status',
    'Gender',
    'Department',
    'Designation',
    'Joining Date',
    'Gross Salary',  # Adjusted from 'Salary' to 'Gross Salary' based on your df.info output
    'Credit Limit',
    'Is Active'
]

# Create a new DataFrame with only the selected columns
filtered_df = df[columns_to_keep]

# Display the filtered DataFrame
print(filtered_df.info())
```

```python
import pandas as pd

# Ensure 'Joining Date' is in datetime format
filtered_df['Joining Date'] = pd.to_datetime(filtered_df['Joining Date'], errors='coerce')

# Calculate Tenure as the difference between today's date and Joining Date
today = pd.Timestamp('now')
filtered_df['Tenure'] = (today - filtered_df['Joining Date']).dt.days

# Display the updated DataFrame
print(filtered_df.info())
print(filtered_df.head())  # Optional: Display the first few rows to check the new column
```

```python
import pandas as pd

# Ensure 'Joining Date' is in datetime format
filtered_df['Joining Date'] = pd.to_datetime(filtered_df['Joining Date'], errors='coerce')

# Calculate Tenure in months
today = pd.Timestamp('now')
filtered_df['Tenure'] = (today - filtered_df['Joining Date']) / pd.Timedelta(days=30)

# Drop rows with NaN in the 'Tenure' column
filtered_df.dropna(subset=['Tenure'], inplace=True)

# Convert Tenure to integer
filtered_df['Tenure'] = filtered_df['Tenure'].astype(int)

# Display the updated DataFrame
print(filtered_df.info())
print(filtered_df.head())  # Optional: Display the first few rows to check the new column
```
